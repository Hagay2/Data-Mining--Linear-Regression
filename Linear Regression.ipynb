```python
#Removing the columns with the Nans
taxi.drop(taxi.columns[[0, 3, 4, 7, 14, 17, 18, 19 ]], axis=1) 
```

```python
#Separting the Features from the labels, Separating X from y 
X = taxi[['PULocationID', 'DOLocationID', 'trip_distance', 'fare_amount', 'extra', 'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge']]
X
```

```python
#extracting the labels
y = taxi[['total_amount']]
y
```

```python
#Split your data into training and testing sets

from sklearn.model_selection import train_test_split

#Splitting the dataset into the training and test sets - 70% of the data is used for training and 30% is used for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 10)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

#Note here we are calling our model 'classifier.' In the titantic example, we called our model lr. This is your choice. 
classifier = LinearRegression()

#Note the use of .vales and .ravel() 
#.values is used to extract the dataframe values to a numpy array

#ravel() is used to change a 2-dimensional array or a multi-dimensional array into a contiguous flattened array.
#The returned array has the same data type as the source array or input array.
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
#Calculating the coefficients
coefficients = classifier.coef_
print(f'Coefficients {coefficients}')
```

```python
#Calculating R Squared
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f'R2: {r2}')
```

```python
#Calculating Mean Absolute Error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print(f'MAE {mae}')
```

```python
#Calculating Mean Squared Error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(f'MSE: {mse}')
```

```python
#Calculating Root Mean Squared Error
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f'RMSE: {rmse}')
```
